```python
#cada seccion esta configurada de tal manera que si se comete un error vuelve aparecer la pregunta con tal de que sea correctamente contestada

def obtener_numero_estudiantes():
  #Aqui el codigo busca cuantos estudiantes son y si ingresa una letra, escribe error y repite la pregunta
    while True:
        try:
            num_estudiantes = int(input("¿Cuántos estudiantes son? "))
            return num_estudiantes
        except ValueError:
            print("Entrada inválida. Por favor, ingresa un número.")

def obtener_nombre_estudiante():
  #Se da el nombre del estudiante y si se da un numero repite la pregunta
    while True:
        nombre = input("¿Cuál es el nombre del estudiante? ")
        if not nombre.strip():
            print("El nombre no puede estar vacío. Intenta de nuevo.")
        elif not nombre.replace(" ", "").isalpha(): #lo puse para ver que tenga letra y espacio
            print("El nombre solo puede contener letras. Intenta de nuevo.")
        else:
            return nombre

def obtener_numero_asignaturas():
  #se da el numero de asignaturas y si no esta correcto se repite la pregunta
    while True:
        try:
            num_asignaturas = int(input("¿Cuántas asignaturas hay? "))
            return num_asignaturas
        except ValueError:
            print("Entrada inválida. Por favor, ingresa un número.")


def obtener_asignaturas_y_calificaciones(num_asignaturas):
    # con el numero de asignaturas damos el nombre de la asignatura y la calificacion correspondiente
    asignaturas = {}
    for i in range(num_asignaturas):
        # Primero, obtenemos el nombre y nos aseguramos que este escrito con letras
        while True:
            nombre = input(f"Ingresa el nombre de la asignatura #{i + 1}: ")
            if not nombre.strip():
                print("El nombre de la asignatura no puede estar vacío.")
            elif not nombre.replace(" ", "").isalpha():
                print("El nombre solo puede contener letras. Intenta de nuevo.")
            else:
                break

        # Luego, obtenemos la calificación para ese nombre, por ejemplo matematicas:8.0
        while True:
            try:
                nota = float(input(f"Ingresa la calificación para '{nombre}': "))
                asignaturas[nombre] = nota
                break
            except ValueError:
                print("Calificación inválida. Por favor, ingresa un número.")
    return asignaturas # Devolvemos el diccionario completo

def calcular_promedio(calificaciones):
    # La función ahora recibe una lista de valores para calcular el promedio
    if not calificaciones:
        return 0
    return sum(calificaciones) / len(calificaciones)

def determinar_estado(promedio):
  #en el promedio usamos de 6.0 como nota de aprobacion hasta maximo 10.0
    if promedio >= 6.0 and promedio <= 10.0:
        return "Aprobado"
    else:
        return "Reprobado"

def imprimir_resumen(estudiantes):
    if not estudiantes:
        print("No hay datos de estudiantes para mostrar.")
        return
# modifique el esqueleto para que me apareciera promedio y asignatura con su nota correspondiente
    print("\n--- Resumen de estudiantes ---")
    for estudiante in estudiantes:
        print(f"Estudiante: {estudiante['nombre']}, Promedio: {estudiante['promedio']:.2f}, Estado: {estudiante['estado']}")
        print("Asignaturas y Calificaciones:")
        for nombre, nota in estudiante['asignaturas'].items(): # Iteramos sobre el diccionario de asignaturas
            print(f"  - {nombre}: {nota}")
        print("------------------------------") #esto es para hacer una linea divisora entre el el estado del estudiante y sus calificaciones por asignatura


num_estudiantes = obtener_numero_estudiantes()
estudiantes = []

for _ in range(num_estudiantes):
    nombre_estudiante = obtener_nombre_estudiante()
    num_asignaturas = obtener_numero_asignaturas()

    #  Aquí tenemos una sola asignatura
    asignaturas_y_calificaciones = obtener_asignaturas_y_calificaciones(num_asignaturas)

    # Obtenemos solo los valores para calcular el promedio
    promedio = calcular_promedio(list(asignaturas_y_calificaciones.values()))
    estado = determinar_estado(promedio)

    estudiantes.append({
        'nombre': nombre_estudiante,
        'asignaturas': asignaturas_y_calificaciones, # Guardamos el cambio aqui
        'promedio': promedio,
        'estado': estado
    })

imprimir_resumen(estudiantes)
```
